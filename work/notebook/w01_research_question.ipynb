# Week 1 – Research Question

## Lane 4: CTR / Engagement Opportunity Scoring



### Problem statement

Many webpages receive high search visibility but attract relatively few clicks and show weak user engagement. Identifying these pages manually is difficult, especially for large content inventories. This project seeks to use observable search and engagement metrics to identify and prioritize pages with the greatest optimization potential, helping content teams focus their efforts on the pages most likely to benefit from improvements.

# Research Question
Which high-visibility pages have both low click-through rates and weak user engagement, indicating the greatest opportunity for optimization?

#Why?
Optimizing pages with high visibility but poor performance can increase website traffic without requiring higher search rankings.

A ranked list of improvement opportunities helps content teams focus their efforts where changes are likely to have the greatest impact.

#Decision and action

## Decision and Action

**Decision**

Which high-visibility pages should the content team review first?

**Action**

Generate a ranked list of pages that should be reviewed for metadata improvements, content updates, or user experience enhancements based on CTR and engagement signals.

#**Quick** **look at the data**


The starter dataset contains **30,000 content pages** with observable search and engagement signals.

Search visibility varies considerably across pages. While the **median page receives 731 impressions over 90 days**, the average is **5,200 impressions**, indicating a highly skewed distribution where a relatively small number of pages receive very high visibility.

User interaction also varies substantially. The **median CTR is only 0.07%**, while the **median page receives 7 sessions over 90 days**. This variation suggests that not all visible pages convert search impressions into clicks or meaningful user engagement, motivating a data-driven approach to identify pages with the greatest optimization potential.



#What i can claim based on the dataset

This project can identify pages that appear to have higher optimization potential based on observable search and engagement signals such as impressions, CTR, sessions, and engagement metrics.

**The final output will be a ranked list of pages that may deserve review for metadata, content, or user experience improvements.**



#What i can't claim

This analysis cannot prove that updating a page will increase its search ranking, CTR, or user engagement.

It also cannot determine the exact reason why a page has low CTR or weak engagement, as factors such as search intent, competition, seasonality, and user behaviour are not fully captured in the starter dataset.
